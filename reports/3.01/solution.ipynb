# Задание 2

```python
def get_line_segments(line: np.array) -> np.array:
    segments = []
    for i in range(line.shape[1]-1):
        segments.append([line[:, i], line[:,i+1]])
    return np.array(segments)


def find_segment_center(line:np.array)->np.array:
    return (line[1,:]+line[0,:]) / 2


def find_segment_normal(line: np.array)->np.array:
    "Normal with start at segment center and length 1"
    d = line[1, :] - line[0, :]
    normal_d = np.array([-d[1], d[0]])
    normal_d = normal_d / np.sqrt(normal_d @ normal_d)
    return normal_d
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.set_xlabel('$x$, мм', loc='right')
ax.set_ylabel('$y$, мм', loc='top', rotation=0)

ax.set_aspect(1)
ax.set_ylim([0, 200])
ax.set_xlim([0, 280])
ax.grid()


ax.set_yticks(ticks=image1_y)
ax.set_xticks(ticks=np.arange(20, 280+1, 20))

for i in range(len(image1_points)):
    xs = image1_points[i, 0, :]
    ys = image1_points[i, 1, :]

    line_start_x = xs[0]
    line_start_y = ys[0]
    annotation_y_offset = -12 - int(i % 2)*5
    ax.annotate(
        f"$\\varphi={image2_phi[i]}$В", 
        xy=(line_start_x, line_start_y), xytext=(-15, annotation_y_offset), textcoords='offset points'
    )
    
    ax.plot(xs, ys, marker='o', linewidth=.75, linestyle='--', color='black', markersize=2)

    for segment in get_line_segments(image1_points[i]):
        center = find_segment_center(segment)
        ax.arrow(*center, *(find_segment_normal(segment)*20), color='black', width=.5, linewidth=0, head_width=2)
    
plt.show(fig)
fig.savefig('img/task1-arrows.svg', format='svg', bbox_inches='tight')
```
